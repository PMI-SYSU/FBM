```python
import numpy as np
import matplotlib.pyplot as plt
```

---

# Figure 0

```python
fermi_surface=0.45
lam=0.01
```

```python
distance_bose_row_mini_T = np.loadtxt("FBM/data/distance/distance_bose_dF=0.455_lam=0.01_0.csv")
distance_fermi_row_mini_T = np.loadtxt("FBM/data/distance/distance_fermi_dF=0.455_lam=0.01_0.csv")
```

```python
distance_bose_row_mini_T=distance_bose_row_mini_T.T
distance_fermi_row_mini_T=distance_fermi_row_mini_T.T
```

```python
len(distance_fermi_row_mini_T)
```

```python
distance_bose = tuple(distance_bose_row_mini_T[i][j] for j in range(100) for i in range(20))
distance_fermi = tuple(distance_fermi_row_mini_T[i][j] for j in range(100) for i in range(20))

x = list(range(len(distance_bose)))
```

```python
# 

fig = plt.figure(figsize=(4,3), dpi=500)


ax1 = fig.add_subplot(111)
ax1.plot(x,distance_fermi,color='#FFA500',label='Fermion Pair')
ax1.plot(x,distance_bose,'#228B22',label='Boson Pair')

ax1.plot(x, [fermi_surface] * len(x), color='black', linestyle='dashdot', linewidth=1.5 )
plt.text(590, fermi_surface + 0.04, r'$d_F=$0.455', ha='right', va='top', color='black', fontsize=11)

ax1.yaxis.set_label_position('left')
ax1.yaxis.tick_left()

ax1.set_xlim(-40,1000)
ax1.set_ylim(-0.03,0.511)

plt.xlabel("Minibatch",fontsize=15)
plt.ylabel("Intra-pair Distance",fontsize=15)
plt.xticks([0,250,500,750,1000],fontsize=10)
plt.yticks([0.0,0.25,0.5],fontsize=10)
# plt.tight_layout()
ax1.yaxis.set_label_position('left')
ax1.yaxis.tick_left()

legend = ax1.legend(loc='center right',fontsize=10)

frame = legend.get_frame()
frame.set_alpha(0.7)
plt.savefig('2',bbox_inches = 'tight')

# 
plt.show()
```
