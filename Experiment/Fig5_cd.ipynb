```python
import matplotlib.pyplot as plt
import numpy as np
start = 0
end = 1.2
interval = 0.01
x_coords = [round(start + i * interval, 3) for i in range(int((end - start) / interval) + 1)]

area_FGSM = np.load('FBM/data/acc/area_FGSM.npy')
area_WN = np.load('FBM/data/acc/area_WN.npy')
C = ['#100150','#4D028F','#B0007F','#EA2462', '#FF7449','#FFB946']
x_star = [0.1, 0.2,0.375, 0.465, 0.56, 1.0]

y1_star = [area_FGSM[9], area_FGSM[19], area_FGSM[37], area_FGSM[46], area_FGSM[55], area_FGSM[99]]
y2_star = [area_WN[9], area_WN[19], area_WN[37], area_WN[46], area_WN[55], area_WN[99]]



plt.figure(figsize=(4, 3), dpi=300)
plt.plot(x_coords, area_FGSM,  marker='s',ms=4, linestyle='-', color='#858597',label='FBM')

for i in range(len(C)):
    plt.plot(x_star[i], y1_star[i],  marker='*',ms=12, linestyle='', color=C[i])


plt.hlines(y=4.288, xmin=-0.05, xmax=1.21, linestyle='--', linewidth=3, color='#CFB0E8', label='MLP')
#plt.title('Area under FBM Curves up to 0.8')
plt.xlabel(r'$d_F$', fontsize=15)
plt.ylabel('$\mathcal{R}_{FGSM}$', fontsize=15)
plt.legend(fontsize=12)

# plt.savefig('FGSM Integral.png', bbox_inches='tight')
plt.show()



plt.figure(figsize=(4, 3), dpi=300)
plt.plot(x_coords, area_WN,  marker='s',ms=4, linestyle='-', color='#858597',label='FBM')
for i in range(len(C)):
    plt.plot(x_star[i], y2_star[i],  marker='*',ms=12, linestyle='', color=C[i])

plt.hlines(y=41.19283333333334, xmin=-0.05, xmax=1.21, linestyle='--', linewidth=3, color='#CFB0E8', label='MLP', zorder=0)

#plt.title('Area under FBM Curves up to 0.8')
plt.xlabel(r'$d_F$', fontsize=15)
plt.ylabel('$\mathcal{R}_{Noise}$', fontsize=15)
plt.legend(fontsize=12)

# plt.savefig('Gaussian Integral.png', bbox_inches='tight')
plt.show()
```
